```python
# Ativação da biblioteca pandas:

import pandas as pd
```

```python
# DataFrame:

df = pd.read_csv('D:/yanna/dengue/DENGBR15.csv')
```

```python
# Carregando o DataFrame para verficicar se a tabela foi lida de forma integral:
## df.head() -> mostra as 5 primeiras linhas do DataFrame, caso coloque algum número entre (), este será o novo numero de linhas
## df.tail() -> mostra as 5 ultimas linhas do DataFrame, caso coloque algum número entre (), este será o novo numero de linhas

df.head()
```

```python
# Demonstra a quantidade de memoria consumida em bytes:

memory = df.memory_usage(deep = True)
```

```python
print(memory)
```

```python
# Transformando bytes em Gbytes

memory.sum() / (1024**3)
```

```python
# df.info() -> demonstra o Dtype de cada uma das colunas, sendo object utilizado para dados do tipo string(str)ou não reconhecidos pelo Python

df.info('ID_AGRAVO')
```

```python
# loc -> funciona para selecionarmos, através de uma lista, somente as colunas que serão utilizadas.
## o numero colocado logo após o primeiro [ indica o índice inicial e o indicado após os : indica o índice final, caso não haja, significa que serão considerados todos os indices após o inicialmente indicado.
### Para evitarmos de digitar todo o Dataframe, atribuimos uma nova variável (dengue15) com todos os parametros setados.

dengue15 = df.loc[0:, ['NU_ANO', 'SG_UF_NOT', 'ID_MUNICIP', 'MUNICIPIO', 'UF', 'DT_OBITO']]
```

```python
dengue15.info()
```

```python
print(dengue15)
```

```python
dengue15.set_index(['NU_ANO'], inplace = True)
```

```python
# fillna() -> transforma todos os NA informados nas colunas no número colocado entre ()
## sempre que o parametro inplace=True for utilizado, a base de dados original será modificada.

dengue15.fillna(0, inplace=True)
```

```python
# Salvando a base de dados tratada em um novo arquivo .csv

dengue15.to_csv('DENG15_TRATADO.csv')
```
